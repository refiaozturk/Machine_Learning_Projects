**Dataset: Crime Statistics by U.S. States**

This dataset contains **crime rates of various U.S. states** and can be used to determine security policies for the states. Crime rates are analyzed to understand the overall security situation of the states and classify them as **high, medium, or low crime rate** states.

---

**Features in the Dataset:**

| **Feature**   | **Description**                                           |
|---------------|-----------------------------------------------------------|
| **Murder**    | The murder (homicide) rate per 100,000 people.            |
| **Assault**   | The number of assault crimes per 100,000 people.          |
| **UrbanPop**  | The percentage (%) of the state's urban population.       |
| **Rape**      | The number of rape crimes per 100,000 people.             |

---

**Scenario: Determining Security Policies**

- **Objective:**  
  By grouping states based on crime rates (such as high, medium, low), develop security measures and policies tailored to each group.

- **Example Situation:**
  - States with high crime rates may be assigned more security personnel.
  - States with medium crime rates may implement crime prevention programs.
  - States with low crime rates may maintain the current security infrastructure.

- **Outcome:**  
  The clusters obtained will serve as a guide to optimize security and crime policies in the states.

---

**Analysis Plan: Classifying States through Clustering**

1. **Data Preparation:**
   - Since the features have different units, the data should be **scaled**.
   - For example, "Murder" and "Rape" rates are direct numerical values, whereas "UrbanPop" is a percentage. Hence, scaling is important.

2. **Clustering Methods:**
   - **K-Means Clustering:** 
     - A fast and effective method to group states based on crime rates.
   - **Hierarchical Clustering:**
     - The clustering structure can be visualized using a dendrogram, and the number of clusters can be determined.

3. **Clustering Criteria:**
   - Three clusters based on crime rates: **high, medium, low** crime rates could be created.

---

**Outcome**

- **Obtained Clusters:**
  1. **High Crime Rate:** States with high murder, assault, and rape rates.
  2. **Medium Crime Rate:** States with an average level of crime rates.
  3. **Low Crime Rate:** States with low crime rates and relatively safer.

- **Action Plan:**
  - **High Crime Rate:** Increase the number of security personnel, implement crime prevention policies.
  - **Medium Crime Rate:** Take preventive measures to prevent further crime rate increases.
  - **Low Crime Rate:** Maintain the current security policies.

# <font color='mediumseagreen'> <b>Hopkins Test</b><font color='black'>  

```python
from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd

def hopkins(data_frame, sampling_size):
    
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    
    # Get the distance to their nearest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    
    # Get the distance to their nearest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]
```

# <font color='mediumseagreen'> <b>Dendrogram</b><font color='black'>  

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
hc_ward = linkage(y = X, method = "ward")
hc_complete = linkage(X, "complete")
hc_average = linkage(X, "average")
hc_single = linkage(X, "single")
```

```python
plt.figure(figsize = (14,7))
dendrogram(hc_ward, leaf_font_size = 10);
```

```python
# simpler dendrogram drawing

plt.figure(figsize = (14,7))
dendrogram(hc_ward,
           truncate_mode = "lastp",
           p = 10,
           show_contracted = True,   
           leaf_font_size = 10);

# show_contracted = True: The junction heights of clusters that are truncated (not shown) in the dendrogram are indicated by horizontal lines.
```

```python
plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, truncate_mode = "lastp", p = 10, show_contracted = True, leaf_font_size = 10);
```

# <font color='mediumseagreen'> <b>Agglomerative Clustering</b><font color='black'>  

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
K = range(2,11)

for k in K:
    model = AgglomerativeClustering(n_clusters = k)
    model.fit_predict(X)
    print(f'Silhouette Score for {k} clusters: {silhouette_score(df, model.labels_)}')
```

```python
X.head()
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage = "ward")
clusters = model.fit_predict(X)
X["cluster"] = clusters
```
