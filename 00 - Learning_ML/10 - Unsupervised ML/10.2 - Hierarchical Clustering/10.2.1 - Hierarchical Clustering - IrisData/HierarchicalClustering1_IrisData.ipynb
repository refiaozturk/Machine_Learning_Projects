**Hierarchical Clustering**

**Differences Between K-Means and Hierarchical Clustering and the Use of Dendrograms**

Traditionally, in unsupervised algorithms, clustering decisions are often made based on a single metric such as the **Elbow Method**. However, this approach can sometimes yield unreliable results due to its reliance on a single criterion. **In this training, we used multiple metrics (inertia, silhouette score, visual analysis, elbow method) to determine the optimal number of clusters for the K-Means algorithm.** This approach provides a more reliable and comprehensive evaluation.

---

**K-Means vs Hierarchical Clustering Approach**

- **K-Means:**  
  - Requires defining a model before clustering begins, even with default parameters.  
  - Metrics like **Elbow Method** and **Silhouette Score** are used to determine the number of clusters.  
  - Clustering principle:  
    - **Minimize distance** within clusters.  
    - **Maximize distance** between clusters.

- **Hierarchical Clustering:**  
  - A **dendrogram** is initially created to understand the data structure without defining a model.  
  - It groups data points based on a hierarchical structure, either merging small groups into larger ones or splitting large groups iteratively. This process is visualized using a **dendrogram**.  
  - **Dendrograms** illustrate how data points are hierarchically grouped and help decide the number of clusters.  
  - Approach:  
    - **Agglomerative (Bottom-Up):** Starts with individual data points and merges them.  
    - **Divisive (Top-Down):** Starts with a large cluster and splits it into smaller clusters.

---

**Types of Hierarchical Clustering**

1. **Agglomerative Clustering:**
   - Uses a **"bottom-up" approach**.
   - Initially considers each data point as an individual cluster, merging similar ones iteratively.
   - Continues until all data points form a single cluster.
   - **More commonly used method.**

2. **Divisive Clustering:**
   - Uses a **"top-down" approach**.
   - Starts with all data points in one cluster, dividing them into smaller clusters iteratively.
   - Less commonly used.

---

**Linkage Methods in Hierarchical Clustering**
During hierarchical clustering, linkage criteria are used to decide which clusters to merge. Common linkage methods include:

1. **Single Linkage:**
   - Considers the shortest distance between data points in two clusters.

2. **Complete Linkage:**
   - Considers the longest distance between data points in two clusters.

3. **Average Linkage:**
   - Uses the average distance between all data points in two clusters.

4. **Ward's Linkage:**
   - Minimizes the total variance resulting from merging clusters. Works with Euclidean distance and forms compact clusters.

---

**How Hierarchical Clustering Works**

**Steps for Agglomerative Clustering:**

1. **Initialization:**
   - Treat each data point as its own cluster.

2. **Similarity Calculation:**
   - Measure the similarity (or distance) between clusters. Common metrics include:
     - **Euclidean Distance**
     - **Manhattan Distance**
     - **Cosine Similarity**

3. **Cluster Merging:**
   - Merge the two most similar clusters (smallest distance).

4. **Repeat Process:**
   - Continue until all data points form a single cluster.

5. **Final Output:**
   - Visualize the results using a dendrogram. The dendrogram illustrates the hierarchical structure of clusters.

---

**Advantages**

1. **Dendrogram Visualization:**
   - Provides a visual representation of the clustering structure.
   - Helps understand the hierarchical organization of data.

2. **No Predefined Cluster Number:**
   - The number of clusters is not predetermined; it can be decided based on the dendrogram.

3. **Supports Various Distance Metrics:**
   - Works with metrics like Euclidean, Manhattan, etc.

---

**Disadvantages**

1. **Computational Complexity:**
   - High computational cost for large datasets (O(nÂ²)).
   - Agglomerative clustering becomes slower as the number of data points increases.

2. **Error Propagation:**
   - A small misstep in merging can affect the entire hierarchy.

3. **Lack of Flexibility:**
   - Results are highly sensitive to the choice of linkage and distance metric.

---

**Comparison Between Hierarchical Clustering and K-Means**

| **Feature**                  | **K-Means**                                    | **Hierarchical Clustering**                  |
|------------------------------|-----------------------------------------------|----------------------------------------------|
| **Model Definition**          | Requires model definition.                   | Uses dendrogram before defining a model.      |
| **Cluster Number Determination** | Determined using metrics like Elbow or Silhouette. | Decided visually from the dendrogram.         |
| **Approach**                  | Predefined number of clusters is required.   | Number of clusters can be determined later.   |
| **Clustering Principle**      | Minimizes intra-cluster distance and maximizes inter-cluster distance. | Similar principles, but in a hierarchical structure. |
| **Speed and Scalability**     | More efficient for large datasets.           | Computationally intensive for large datasets. |

---

**Application Areas**

1. **Genetics and Biology:** Analyzing DNA sequences and classifying gene clusters.
2. **Marketing:** Customer segmentation.
3. **Social Network Analysis:** Understanding user groups.
4. **Document Clustering:** Grouping similar text documents.

# <font color='mediumpurple'> <b>Scaling Data</b><font color='black'>  

**Scaling in Hierarchical Clustering**

Hierarchical clustering algorithms, especially when using distance measures like **Euclidean distance**, are sensitive to the scales of variables. If there are significant scale differences between features, those with larger values can dominate others and distort clustering results. **Therefore, scaling is often crucial.**

---

1. **When Scaling is Not Applied:**
   - Features with large scales influence the distances between clusters, overshadowing the effects of smaller-scale features.
   - As a result, clusters may deviate from the natural structure of the data.

2. **When Scaling is Applied:**
   - All features are brought to the same scale, allowing the algorithm to consider all features equally.
   - This leads to more balanced and meaningful clusters.

---

**Is Scaling Necessary for the Iris Dataset?**

When working with the **Iris dataset**:
- All features are measured in centimeters (e.g., "sepal_length", "petal_width").
- Since the scales of the features are already similar (e.g., within the range of 0-10), **additional scaling is usually not required.**
- However, for other datasets or when features have different units, scaling plays a critical role.

# <font color='mediumpurple'> <b>Hopkins Test</b><font color='black'>  

**What is the Hopkins Test?**

The Hopkins test evaluates whether a dataset follows a uniform (random) distribution and determines its suitability for clustering analysis.

- **Result Interpretation:**
  - If the Hopkins test result is **below 0.5**, the dataset is suitable for clustering.
  - Values of **0.3 or lower** strongly indicate that the dataset has a meaningful cluster structure.

---

**In Summary:**
The Hopkins test is a critical metric for assessing the suitability of a dataset for clustering analysis. A test result of 0.3 or lower suggests that the dataset likely has a naturally clustered structure.

````python
from sklearn.neighbors import BallTree
import numpy as np
import pandas as pd

def hopkins(data_frame, sampling_size):
    
    """Assess the clusterability of a dataset. A score between 0 and 1, a score around 0.5 express
    no clusterability and a score tending to 0 express a high cluster tendency.
    Parameters
    ----------
    data_frame : numpy array
        The input dataset
    sampling_size : int
        The sampling size which is used to evaluate the number of DataFrame.
    Returns
    ---------------------
    score : float
        The hopkins score of the dataset (between 0 and 1)
    Examples
    --------
    >>> from sklearn import datasets
    >>> from pyclustertend import hopkins
    >>> X = datasets.load_iris().data
    >>> hopkins(X,150)
    0.16
    """
    
    if type(data_frame) == np.ndarray:
        data_frame = pd.DataFrame(data_frame)
    
    # Sample n observations from D : P
    if sampling_size > data_frame.shape[0]:
        raise Exception(
            'The number of sample of sample is bigger than the shape of D')
    data_frame_sample = data_frame.sample(n=sampling_size)
    
    # Get the distance to their nearest neighbors in D : X
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(data_frame_sample, k=2)
    data_frame_sample_distances_to_nearest_neighbours = dist[:, 1]
    
    # Randomly simulate n points with the same variation as in D : Q.
    max_data_frame = data_frame.max()
    min_data_frame = data_frame.min()
    uniformly_selected_values_0 = np.random.uniform(min_data_frame[0], max_data_frame[0], sampling_size)
    uniformly_selected_values_1 = np.random.uniform(min_data_frame[1], max_data_frame[1], sampling_size)
    uniformly_selected_observations = np.column_stack((uniformly_selected_values_0, uniformly_selected_values_1))
    if len(max_data_frame) >= 2:
        for i in range(2, len(max_data_frame)):
            uniformly_selected_values_i = np.random.uniform(min_data_frame[i], max_data_frame[i], sampling_size)
            to_stack = (uniformly_selected_observations, uniformly_selected_values_i)
            uniformly_selected_observations = np.column_stack(to_stack)
    uniformly_selected_observations_df = pd.DataFrame(uniformly_selected_observations)
    
    # Get the distance to their nearest neighbors in D : Y
    tree = BallTree(data_frame, leaf_size=2)
    dist, _ = tree.query(uniformly_selected_observations_df, k=1)
    uniformly_df_distances_to_nearest_neighbours = dist
    
    # return the hopkins score
    x = sum(data_frame_sample_distances_to_nearest_neighbours)
    y = sum(uniformly_df_distances_to_nearest_neighbours)
    if x + y == 0:
        raise Exception('The denominator of the hopkins statistics is null')
    return x / (x + y)[0]

# normally this function is available in pyclustertend's hopkins library. However, the function itself will be used directly against any errors that may occur. 
# Therefore, pyclustertend does not need to be installed!
````

# <font color='mediumpurple'> <b>Dendrogram & Linkage</b><font color='black'>  

**What is a Dendrogram and How is it Used?**

A **dendrogram** is a **tree-like diagram** that visualizes the results of a hierarchical clustering model.

- **Function:**  
  - Shows the order in which data points are merged.  
  - Indicates the distance levels at which these merges occur.  
  - Illustrates how all data points eventually merge into a single cluster.

- **Determining the Number of Clusters:**  
  1. **Observe the branching points on the dendrogram.**  
  2. **Identify the largest vertical gap (distance).**  
  3. **Draw a horizontal line through the midpoint of this gap.**  
  4. The number of branches (lines) intersected by this line represents the optimal number of clusters.

---

**Principles of Hierarchical Clustering**
- **Intra-cluster Cohesion:**  
  - Observations within a cluster should be tightly connected. (Minimum distance)
- **Inter-cluster Separation:**  
  - Different clusters should be as distinct from each other as possible. (Maximum distance)
- **No Need for Model Definition:**  
  - Using a dendrogram, the ideal number of clusters can be visually determined before defining a model.

---

**Dendrogram Function: Parameters and Their Meanings**

The **dendrogram** function is used to visualize the results of hierarchical clustering. It displays how data points are grouped and the hierarchical structure of this grouping in a tree diagram (dendrogram). Below are some important parameters of this function and their meanings:

---

**Key Parameters:**

1. **Z (Linkage Matrix):**  
   - **Definition:** The linkage matrix containing clustering results.  
   - **Source:** Generated by the **linkage** function, which holds information about agglomerative clustering.  
   - **Example:**  
     ```python
     from scipy.cluster.hierarchy import linkage
     Z = linkage(X, method='ward')
     ```

2. **p (Cut Level):**  
   - **Definition:** Specifies the level or cut line on the dendrogram.  
   - If a value is given, the dendrogram is reduced to p levels and shows up to p different colors.

3. **truncate_mode:**  
   - **Definition:** Determines the truncation mode of the dendrogram.  
   - Values:  
     - **lastp**: Shows the last p merged clusters.  
     - **mlab**: Similar to MATLAB's old dendrogram function truncation.

4. **color_threshold:**  
   - **Definition:** Specifies the threshold for determining line colors based on clustering levels.  
   - If set to **None**, the threshold is calculated automatically.  
   - Clusters are displayed in different colors based on the specified threshold.

5. **get_leaves:**  
   - **Definition:** Returns the IDs of leaf nodes.  
   - **Value:** If True, leaf nodes are returned.

6. **orientation:**  
   - **Definition:** Specifies the orientation of the dendrogram.  
   - **Values:** **top**, **bottom**, **left**, **right**.  
   - For example, **top** draws the dendrogram from top to bottom.

7. **labels:**  
   - **Definition:** An array of leaf labels.  
   - **Default:** If not specified, indices are used as leaf labels.

8. **count_sort:**  
   - **Definition:** Sorts leaves by the number of items.  
   - **Values:** **True**, **ascending**, **descending**.

9. **distance_sort:**  
   - **Definition:** Sorts leaves by linkage distances.  
   - **Values:** **True**, **ascending**, **descending**.

10. **show_leaf_counts:**  
    - **Definition:** Adds the number of items in a cluster next to the leaves.  
    - **Value:** If True, the number of items in each leaf is displayed.

11. **no_plot:**  
    - **Definition:** Disables dendrogram plotting.  
    - **Value:** If True, the dendrogram is not drawn, but the necessary information is returned.

12. **no_labels:**  
    - **Definition:** Hides leaf labels.  
    - **Value:** If True, leaf labels are not shown.

13. **leaf_rotation:**  
    - **Definition:** Specifies the rotation angle of leaf labels.  
    - **Default:** Typically set to 90 degrees.

14. **leaf_font_size:**  
    - **Definition:** Determines the font size of leaf labels.

15. **show_contracted:**  
    - **Definition:** Displays contracted cluster heights with lines.  
    - **Value:** If True, these details are shown.

---

**Usage Example:**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Dataset and connection matrix
Z = linkage(X, method='ward')

# Plotting Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(
    Z,
    truncate_mode='lastp',  # Show the last p clusters
    p=5,                    # Display 5 different clusters
    color_threshold=1.5,    # Color changes at the 1.5 level
    leaf_rotation=90,       # Rotate leaf labels by 90 degrees
    leaf_font_size=10,      # Font size of leaf labels
    show_leaf_counts=True   # Add the number of items in clusters next to the leaves

)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

**Linkage Function: Hierarchical Clustering and Linkage Methods**

The **linkage** function performs hierarchical clustering. It takes a given dataset and creates a hierarchical clustering structure by merging the two closest clusters at each step. This function uses different linkage methods based on similarity or distance measurements between data points.

---

**Linkage Parameters and Methods**

1. **ward (Most Common Method):**
   - **Description:** Minimizes the total variance within merged clusters.
   - **Purpose:** Creates homogeneous clusters.
   - **Feature:** Works only with Euclidean distance.
   - **Advantage:** Produces tighter, more compact clusters.
   - **Example:** 
     ```python
     hc_ward = linkage(X, method="ward")
     ```

2. **complete (Complete Linkage / Maximum Distance):**
   - **Description:** Considers the distance between the farthest points of two clusters.
   - **Purpose:** Focuses on the maximum separation between clusters.
   - **Feature:** Can form larger clusters.
   - **Example:** 
     ```python
     hc_complete = linkage(X, method="complete")
     ```

3. **average (Average Linkage):**
   - **Description:** Uses the average of all distances between points in two clusters.
   - **Purpose:** Provides an overall measure of distance between clusters.
   - **Feature:** Produces balanced clusters.
   - **Example:** 
     ```python
     hc_average = linkage(X, method="average")
     ```

4. **single (Single Linkage / Minimum Distance):**
   - **Description:** Considers the shortest distance between points in two clusters.
   - **Purpose:** Finds the shortest connection and merges clusters accordingly.
   - **Feature:** May result in long, chain-like clusters (chain effect).
   - **Example:** 
     ```python
     hc_single = linkage(X, method="single")
     ```

---

**Comparison of Linkage Methods**

| **Method**      | **Clustering Logic**                                  | **Advantage**                              | **Disadvantage**                           |
|------------------|-------------------------------------------------------|--------------------------------------------|-------------------------------------------|
| **Ward**        | Minimizes variance, creates tight clusters.            | Compact and homogeneous clusters.          | Works only with Euclidean distance.       |
| **Complete**    | Uses the maximum distance between two clusters.        | Ensures good separation between clusters.  | May create larger clusters.               |
| **Average**     | Uses the average distance between two clusters.        | Produces balanced clusters.                | Can be more general in clustering.        |
| **Single**      | Uses the minimum distance between two clusters.        | Easily merges small and close clusters.    | May form chain-like, elongated clusters.  |

---

**Selection and Use Cases for Linkage Methods**

| **Method**      | **Use Cases and Advantages**                                                                                       | **Disadvantages**                                                                                      | **Example Applications**                                                         |
|------------------|-------------------------------------------------------------------------------------------------------------------|--------------------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------|
| **ward**        | - Minimizes variance, creates compact and homogeneous clusters.                                                   | - Works only with **Euclidean distance**.                                                              | Customer segmentation, genetic data clustering, social media user grouping.      |
|                 | - Suitable when the goal is to form tighter (compact) and balanced clusters.                                      | - May not be effective if clusters have unequal sizes or if variance within the dataset is large.      |                                                                                  |
| **single**      | - Focuses on the shortest distance between clusters. Useful to observe connections between data points.           | - "Chain effect" may result in long, loose clusters.                                                   | Geographic analysis (e.g., map data), social network analysis, water resource clustering. |
| **complete**    | - Uses the maximum distance, ensures wider separation between clusters.                                           | - Noisy datasets or outliers can disrupt results.                                                      | Logistic routes, inter-city distance analysis, product categorization.           |
| **average**     | - A balanced approach using average distance, more resistant to noise and outliers.                              | - Performance may drop with extremely homogeneous or heterogeneous datasets.                           | Marketing campaign targeting, biological data clustering (e.g., protein analysis). |

---

**Detailed Explanations and Example Applications**

**1. Ward (Variance Minimization Method)**
- **Use Case:** Forms compact, homogeneous clusters. Ideal when the clustering goal is to create tight (compact) clusters.
- **Example Applications:**
  - **Customer Segmentation:** Grouping different customer types into homogeneous groups.
  - **Genetic Data Clustering:** Grouping genes based on genetic similarity.
  - **Social Media Analysis:** Dividing users into tightly-knit groups.

---

**2. Single (Minimum Distance Method)**
- **Use Case:** Merges clusters based on the shortest distance. Useful when observing connections between data points.
- **Example Applications:**
  - **Geographic Analysis:** Analyzing connections between settlements on a map.
  - **Social Network Analysis:** Grouping connected nodes in network structures.
  - **Water Resources Analysis:** Grouping nearby rivers or lakes.

---

**3. Complete (Maximum Distance Method)**
- **Use Case:** Measures maximum separation between clusters. Suitable for datasets resistant to noise.
- **Example Applications:**
  - **Logistics Routes:** Grouping cities based on inter-city distances.
  - **Inter-City Distance Analysis:** Creating groups based on distances between the farthest neighborhoods in cities.
  - **Product Categorization:** Categorizing products with significantly different features.

---

**4. Average (Mean Distance Method)**
- **Use Case:** Provides a balanced approach and is resistant to noise and outliers.
- **Example Applications:**
  - **Marketing Campaigns:** Targeting different customer groups based on average distances.
  - **Biological Data Clustering:** Analyzing protein or gene groups.
  - **Educational Analysis:** Clustering students based on their academic performance.

---

**Hierarchical Clustering Process**

1. **Initialization with Data Points:**
   - Each data point starts as a separate cluster.

2. **Similarity Measurement:**
   - The similarity (or distance) between two clusters is calculated based on the selected linkage method.

3. **Merging Clusters:**
   - The two closest (or most suitable) clusters are merged.

4. **Repetition:**
   - The process continues until all clusters merge into one.

5. **Result:**
   - Visualized with a dendrogram, which helps determine the number of clusters.

---

````python
from scipy.cluster.hierarchy import dendrogram, linkage
````

````python
# dendrogram()   

# While dendogram requests data, it also requests information on how distances will be measured.
# For this reason, the linkage function is given to the dendogram by defining the data and specifying how the measurement will be made. 
# In the light of this information, the dendogram assigns observations to clusters and presents these assignment processes to us in the form of a branch.
````

````python
hc_ward = linkage(y=X, method="ward")  # Based on variance - assigns the new point to the cluster where it will least change the current variance.
hc_complete = linkage(X, "complete")   # Based on the farthest point - measures the new point's distance to the farthest points in the existing clusters and assigns it to the closer cluster.
hc_average = linkage(X, "average")     # Based on the average - calculates the average distance of all clusters and measures the new point's distance, assigning it accordingly.
hc_single = linkage(X, "single")       # Based on the closest point - measures the distance of the new point to the closest points in the existing clusters and assigns it to the cluster with the smaller distance.
````

````python
# drawing the dendrogram according to the linkage calculation above

plt.figure(figsize= (20,8))
plt.title("dendrogram")
plt.xlabel("observation")
plt.ylabel("distance")
dendrogram (hc_ward, leaf_font_size=10);  # Making a dendrogram according to the ward from the options above
````

````python
# drawing dendrogram according to all methods

plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size = 10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, leaf_font_size = 10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, leaf_font_size = 10);
````

````python
# simplifying the appearance of graphics

plt.figure(figsize = (20,12))

plt.subplot(221)
plt.title("Ward")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_ward, leaf_font_size = 10, truncate_mode='lastp', p=10)

plt.subplot(222)
plt.title("Complete")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_complete, leaf_font_size = 10, truncate_mode='lastp', p=10)

plt.subplot(223)
plt.title("Average")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_average, leaf_font_size = 10, truncate_mode='lastp', p=10)

plt.subplot(224)
plt.title("Single")
plt.xlabel("Observations")
plt.ylabel("Distance")
dendrogram(hc_single, leaf_font_size = 10, truncate_mode='lastp' , p=10);  # p: number of clusters to form in the last branches

# By setting truncate_mode='lastp' and p=10, the dendogram is simplified to show us only the last 10 clusters.
````
